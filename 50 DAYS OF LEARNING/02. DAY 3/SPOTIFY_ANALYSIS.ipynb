```python
#cleaning data and filling null values
df['album_name']=df['album_name'].fillna("Unknown")
df['album_name']=df['album_name'].fillna("Unknown")
df['track_name']=df['track_name'].fillna("Unknown")
df["duration_ms"]=df["duration_ms"].fillna("Unknown")
df["explicit"]=df["explicit"].fillna("Unknown")
df["danceability"]=df["danceability"].fillna("Unknown")
df["energy"]=df["energy"].fillna("Unknown")
df["key"]=df["key"].fillna("Unknown")
df["loudness"]=df["loudness"].fillna("Unknown")
df['mode']=df['mode'].fillna("Unknown")
df['speechiness']=df['speechiness'].fillna("Unknown")
df['acousticness']=df['acousticness'].fillna("Unknown")
df["tempo"]=df["tempo"].fillna("Unknown")
df["valence"]=df["valence"].fillna("Unknown")
df["track_genre"]=df["track_genre"].fillna("Unknown")
df["liveness"]=df["liveness"].fillna("Unknown")
df['time_signature']=df['time_signature'].fillna("Unknown")
df["instrumentalness"]=df["instrumentalness"].fillna("Unknown")
```

```python
df = df[df["artists"] != "unknown"]
df = df[df["popularity"] != "popularity"]
df["popularity"] = pd.to_numeric(df["popularity"], errors="coerce")
```

```python
#top 10 artists
top_artists = (df.groupby("artists")["popularity"].mean().sort_values(ascending=False).head(10))

top_artists.plot(kind="bar")
plt.xticks(rotation=90)
plt.title("TOP 10 ARTISTS")
plt.xlabel("Artist")
plt.ylabel("Popularity")
plt.show()
```

```python
top_genres = (
    df.groupby('track_genre')['popularity']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,5))

top_genres.plot(kind='bar', color='orange')

plt.title("Top Genres by Average Popularity")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")


plt.show()
```

```python
plt.hist(df["popularity"], bins=30)

plt.title("Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(df['popularity'],df['duration_ms'])
plt.title("Popularity Distribution")
plt.xlabel("Popularity")
plt.ylabel("duration_ms")

plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(
     df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm"
)

plt.title("CORRELATION HEATMAP")
plt.show()
```

```python
explicit=df.groupby('explicit')['popularity'].mean()
explicit.plot(kind='bar',color="green")

plt.title("explicity vs popularity")
plt.xlabel("explicit")
plt.ylabel("popularity")
plt.show()
#
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='duration_ms',y='popularity')

plt.title("duration_ms vs popularity")
plt.xlabel("duration_ms")
plt.ylabel("popularity")
plt.show()
```

```python
sample = df.sample(500)

plt.figure(figsize=(10,5))

plt.scatter(
    sample['danceability'],
    sample['popularity'],
    alpha=0.5
)

plt.title("Danceability vs Popularity")
plt.xlabel("Danceability")
plt.ylabel("Popularity")

plt.grid(True)

plt.show()
```

CONCLUSION:
This analysis explored Spotify track data using Python, Pandas, Matplotlib, and Seaborn to identify patterns in music popularity and audio features. The study revealed that highly popular songs are relatively uncommon, while certain artists consistently achieve stronger popularity scores. Explicit tracks showed slightly higher average popularity, although popularity itself appears to depend on multiple interacting factors rather than a single feature like danceability. Correlation analysis also highlighted meaningful relationships between musical attributes such as energy, loudness, and acousticness.
