```python
df['Outlet_Size']=df['Outlet_Size'].fillna("Unknown")
df['Item_Weight']=df['Item_Weight'].fillna("Unknown")
```

```python
count=df["Outlet_Size"].value_counts()
count.plot(kind="bar",color="green")
plt.title("Outlet Size Analysis")
plt.xlabel("Outlet Size")
plt.ylabel("Count")

plt.show()
```

HENCE,
We can conclude that **Medium sized Item** has higher sales than that of small and high one.

```python
item_type_count=df["Item_Type"].value_counts()
item_type_count.plot(kind="line",color="green")

plt.title="Item Type Analysis"
plt.xlabel("Item Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

HENCE,We can see that **Fruits and Vegetables** has highest sales.

```python
import matplotlib.pyplot as plt

df["Item_Fat_Content"] = df["Item_Fat_Content"].replace({
    "LF": "Low Fat",
    "low fat": "Low Fat",
    "reg": "Regular"
})

comparison = df.groupby("Item_Fat_Content")["Item_Outlet_Sales"].sum()

comparison.plot(kind="bar", color="green")

plt.gca().set_title("Comparison between Item Fat Content and Sales")
plt.xlabel("Item Fat Content")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)

plt.show()
```

HENCE,We can conclude that **Low Fat** Items gets more Sales.

```python
sales=df.groupby("Outlet_Location_Type")["Item_Outlet_Sales"].sum()
vs=sales.plot(kind="bar",color="green")

vs.set_title("Outlet Location Type Analysis")
vs.set_xlabel("Outlet Location Type")
vs.set_ylabel("Total Sales")

plt.show()
```

HENCE,We can conclude that Customers residing in **Tier 3** Location buys more products.
